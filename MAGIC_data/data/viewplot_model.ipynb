```python
y=ecsvdata['nufnu'].data
x=ecsvdata['freq'].data

plt.loglog(x,y,linewidth=1, label='Models '+target)
plt.xlabel('Hz')
plt.ylabel('erg / (cm^2 s)')
plt.xlim(1e9,1e34)
plt.loglog()
plt.grid()
plt.legend(loc='upper right')
plt.show()
```
